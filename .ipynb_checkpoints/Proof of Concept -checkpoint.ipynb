```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

ilk_samaradorlik = 100
osish_foizi = 1
yildagi_kunlar_soni = 365
yildagi_oylar_soni = 365//30
yildagi_haftalar_soni = 365//7


avvalgi_oy_samaradorlik = ilk_samaradorlik #dasturchilarga: shunchaki initialization uchun
avvalgi_hafta_samaradorlik = ilk_samaradorlik #dasturchilarga: shunchaki initialization uchun
avvalgi_kun_samaradorlik = ilk_samaradorlik #dasturchilarga: shunchaki initialization uchun

kunlik_samaradorlik = []
haftalik_samaradorlik = []
oylik_samaradorlik = []

for oy in range(yildagi_oylar_soni):
    ayni_oy_samaradorlik = avvalgi_oy_samaradorlik + avvalgi_oy_samaradorlik*osish_foizi/100.
    oylik_samaradorlik.append(ayni_oy_samaradorlik)
    avvalgi_oy_samaradorlik = ayni_oy_samaradorlik

for hafta in range(yildagi_haftalar_soni):
    ayni_hafta_samaradorlik = avvalgi_hafta_samaradorlik + avvalgi_hafta_samaradorlik*osish_foizi/100.
    haftalik_samaradorlik.append(ayni_hafta_samaradorlik)
    avvalgi_hafta_samaradorlik = ayni_hafta_samaradorlik

for kun in range(yildagi_kunlar_soni):
    ayni_kun_samaradorlik = avvalgi_kun_samaradorlik + avvalgi_kun_samaradorlik*osish_foizi/100.
    kunlik_samaradorlik.append(ayni_kun_samaradorlik)
    avvalgi_kun_samaradorlik = ayni_kun_samaradorlik

         
```

```python
#plot settings
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 15
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size
font = {'size'   : 20}
plt.rc('font', **font)

fig, axs = plt.subplots(2)

axs[0].plot(np.linspace(1,365,yildagi_oylar_soni),oylik_samaradorlik,label = 'har oy',linewidth=4)
axs[0].plot(np.linspace(1,365,yildagi_haftalar_soni),haftalik_samaradorlik,label = 'har hafta',linewidth=4)
axs[0].grid()
axs[0].legend()
axs[0].set_ylabel('Samaradorlik (Maosh)')
axs[0].set_title("Oylik / haftalik / kunlik  1%  o'sish natijasi")

axs[1].plot(np.linspace(1,365,yildagi_oylar_soni),oylik_samaradorlik,label = 'har oy',linewidth=4)
axs[1].plot(np.linspace(1,365,yildagi_haftalar_soni),haftalik_samaradorlik,label = 'har hafta',linewidth=4)
axs[1].plot(np.arange(yildagi_kunlar_soni),kunlik_samaradorlik,label = 'har kun',linewidth=4)
axs[1].grid()
axs[1].legend()
axs[1].set_ylabel('Samaradorlik(Maosh)')

plt.xlabel('1 Yil')
plt.savefig('jahongirga.png',dpi=200)
plt.show()
```
